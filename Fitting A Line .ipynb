```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creating input feature
X = np.linspace(0,1,100)
exp = np.random.choice([2,3])

#creating output feature without noise
y1 = X**exp 
#creating plot
plt.plot(X,y1,'.')
```

```python
#creating data with noise
y = X**exp + np.random.randn(X.shape[0])/10 #this noise is added to account for the errors occured while collecting data and
                                            #depicts the deviations from the true model.

#creating plot
plt.plot(X,y,'.')
```

```python
#creating predictive relationship; using fitting a line

a = np.polyfit(X,y,1)
b = np.poly1d(a)

#plotting
plt.plot(X,y,'.')
plt.plot(X,b(X), label = 'Model')
plt.plot(X,X**exp, label = 'Truth')
plt.legend()
```

```python
#this model worked when the working range was from 0 to 1.
#Hence increasing the working range to examine the model.

X = np.linspace(0,2,100)
exp = np.random.choice([2,3])

y = X**exp + np.random.randn(X.shape[0])/10

p = np.polyfit(X,y,1)
q = np.poly1d(p)
print(q)
print(p)

plt.plot(X,y,'.')
plt.plot(X,q(X),label = 'Model')
plt.plot(X, X**exp, label = 'Truth')
plt.legend()
```

```python
# Model formed is heavily biased. Hence it is required to make the model flexible
#Adding multiple dimensions to our model.


p = np.polyfit(X,y,15)              #returns array of coefficients
q = np.poly1d(p)                    #returns polynomial by taking array of coefficients from earlier step

plt.figure(figsize=[14,6])          #sets the size of the plot, where width=14, and height=6, by default it is [6.8,4.8]

plt.plot(X,y,'.')
plt.plot(X,q(X), label = 'model')   #q(X) means the polynomial Q(x) where the input values are from that of X
plt.legend()
```

```python
#This model not only captured and learned frm the signal, but also from the noise.
#Hence it is overadaptive, and has too many degrees of freedom.
#Main problem is that it will perform badly on unseen data because it has learnt from the noise as well.
```

```python
#Running another iteration with different set

X = np.linspace(0,2.5,100)
y = X**exp + np.random.randn(X.shape[0])/10

plt.figure(figsize=[10,5])
plt.plot(X,q(X), label = 'Model')
plt.plot(X,y,'.', label = '$x^{}$'.format(exp))
plt.legend()
```
